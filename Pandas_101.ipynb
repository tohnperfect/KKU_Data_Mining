## Extra2 Map visualization
https://www.kdnuggets.com/2018/09/visualising-geospatial-data-python-folium.html
https://github.com/python-visualization/folium/blob/master/examples/GeoJSON_and_choropleth.ipynb

```
import folium
import pandas as pd
```

```
data_to_plot = pd.read_csv('worlddata.csv')
```

```
data_to_plot.head()
```

```
country_geo = 'world.json'
```

```
map = folium.Map(location=[100, 0], zoom_start=1.5)
```

```
map
```

```
map.choropleth(geo_data='world.json', data=data_to_plot,
             columns=['CountryName', 'Value'],
             key_on='feature.properties.name',
             fill_color='YlGnBu', fill_opacity=0.7, line_opacity=0.2,
             legend_name='dying age')
```

```
map.save('plot_data.html')
```

```
from IPython.display import HTML
HTML('<iframe src=plot_data.html width=700 height=450></iframe>')
```
